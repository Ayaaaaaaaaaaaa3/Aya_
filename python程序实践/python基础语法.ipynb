## 1.4 数据结构
- ### 列表

```python
# 列表的定义:
# 在Python中，列表（list）是一种用于存储有序集合的数据类型。
# 列表中的元素可以是任何数据类型，并且列表是可变的，这意味着你可以在列表中添加、删除或修改元素。
# 列表的创建:
# 你可以使用方括号 [] 来定义一个列表，并用逗号 , 分隔列表中的元素。以下是一些定义列表的示例：

# 定义一个空列表
empty_list = []

# 定义一个包含整数的列表
numbers = [1, 2, 3, 4, 5]

# 定义一个包含字符串的列表
fruits = ["apple", "banana", "cherry"]

# 定义一个包含不同数据类型的列表
mixed_list = [1, "hello", 3.14, True]

# 定义一个嵌套列表（列表中的元素也是列表）
nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# 访问列表元素
# 你可以使用索引来访问列表中的元素。索引从0开始，负索引表示从列表末尾开始计数。

# 访问列表中的第一个元素
first_fruit = fruits[0]  # "apple"

# 访问列表中的最后一个元素
last_fruit = fruits[-1]  # "cherry"

# 访问嵌套列表中的元素
first_row = nested_list[0]  # [1, 2, 3]
first_element_of_first_row = nested_list[0][0]  # 1

# 修改列表元素
# 你可以通过索引来修改列表中的元素。

# 修改列表中的元素
fruits[1] = "blueberry"
print(fruits)  # ["apple", "blueberry", "cherry"]

# 添加和删除列表元素
# Python提供了多种方法来添加和删除列表中的元素。

# 使用 append() 方法在列表末尾添加元素
fruits.append("orange")
print(fruits)  # ["apple", "blueberry", "cherry", "orange"]

# 使用 insert() 方法在指定位置插入元素
fruits.insert(1, "banana")
print(fruits)  # ["apple", "banana", "blueberry", "cherry", "orange"]

# 使用 remove() 方法删除指定元素
fruits.remove("blueberry")
print(fruits)  # ["apple", "banana", "cherry", "orange"]

# 使用 pop() 方法删除指定位置的元素（默认删除最后一个元素）
last_fruit = fruits.pop()
print(last_fruit)  # "orange"
print(fruits)  # ["apple", "banana", "cherry"]

# 使用 del 关键字删除指定位置的元素
del fruits[1]
print(fruits)  # ["apple", "cherry"]

# 使用 clear() 方法清空列表
fruits.clear()
print(fruits)  # []

列表的常用方法
Python的列表提供了许多内置方法，以下是一些常用的方法：

# 定义一个列表
numbers = [1, 2, 3, 4, 5]

# 使用 len() 函数获取列表的长度
length = len(numbers)  # 5

# 使用 sort() 方法对列表进行排序
numbers.sort()
print(numbers)  # [1, 2, 3, 4, 5]

# 使用 reverse() 方法反转列表
numbers.reverse()
print(numbers)  # [5, 4, 3, 2, 1]

# 使用 index() 方法查找元素的索引
index_of_three = numbers.index(3)  # 2

# 使用 count() 方法统计元素出现的次数
count_of_two = numbers.count(2)  # 1

# 使用 copy() 方法复制列表
numbers_copy = numbers.copy()
print(numbers_copy)  # [5, 4, 3, 2, 1]
```

- ### 元组

```python
# 元组的定义
# 在Python中，元组（tuple）是一种用于存储多个元素的有序集合。
# 与列表不同，元组是不可变的，这意味着一旦创建，元组中的元素就不能被修改、添加或删除。
# 元组通常用于存储一组相关的数据，且这些数据在整个程序的生命周期中不会改变。
# 元组的创建
# 你可以使用圆括号 () 来定义一个元组，并用逗号 , 分隔元组中的元素。以下是一些定义元组的示例：

# 定义一个空元组
empty_tuple = ()

# 定义一个包含整数的元组
numbers = (1, 2, 3, 4, 5)

# 定义一个包含字符串的元组
fruits = ("apple", "banana", "cherry")

# 定义一个包含不同数据类型的元组
mixed_tuple = (1, "hello", 3.14, True)

# 定义一个嵌套元组（元组中的元素也是元组）
nested_tuple = ((1, 2, 3), (4, 5, 6), (7, 8, 9))

# 定义一个只有一个元素的元组（注意逗号）
single_element_tuple = (42,)

# 访问元组元素
# 你可以使用索引来访问元组中的元素。索引从0开始，负索引表示从元组末尾开始计数。

# 访问元组中的第一个元素
first_fruit = fruits[0]  # "apple"

# 访问元组中的最后一个元素
last_fruit = fruits[-1]  # "cherry"

# 访问嵌套元组中的元素
first_row = nested_tuple[0]  # (1, 2, 3)
first_element_of_first_row = nested_tuple[0][0]  # 1

# 元组的不可变性
# 元组是不可变的，这意味着你不能修改、添加或删除元组中的元素。如果尝试修改元组中的元素，会引发错误。

# 尝试修改元组中的元素（会引发错误）
# fruits[1] = "blueberry"  # TypeError: 'tuple' object does not support item assignment

# 尝试删除元组中的元素（会引发错误）
# del fruits[1]  # TypeError: 'tuple' object doesn't support item deletion

# 元组的常用操作
# 尽管元组是不可变的，但你仍然可以进行一些常见的操作，如连接、重复和切片。

# 定义两个元组
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

# 连接元组
concatenated_tuple = tuple1 + tuple2
print(concatenated_tuple)  # (1, 2, 3, 4, 5, 6)

# 重复元组
repeated_tuple = tuple1 * 2
print(repeated_tuple)  # (1, 2, 3, 1, 2, 3)

# 切片元组
sliced_tuple = concatenated_tuple[1:4]
print(sliced_tuple)  # (2, 3, 4)

# 元组的常用方法
# Python的元组提供了一些内置方法，以下是一些常用的方法：

# 定义一个元组
numbers = (1, 2, 3, 4, 5)

# 使用 len() 函数获取元组的长度
length = len(numbers)  # 5

# 使用 index() 方法查找元素的索引
index_of_three = numbers.index(3)  # 2

# 使用 count() 方法统计元素出现的次数
count_of_two = numbers.count(2)  # 1

# 元组的解包
# 元组支持解包操作，可以将元组中的元素直接赋值给多个变量。

# 定义一个元组
person = ("Alice", 30, "Engineer")

# 解包元组
name, age, profession = person
print(name)  # "Alice"
print(age)  # 30
print(profession)  # "Engineer"
```

- ### 字符串

```python
# 字符串的定义
# 在Python中，字符串（string）是一种用于表示文本数据的序列。
# 字符串是不可变的，这意味着一旦创建，字符串中的字符就不能被修改。
# 字符串可以包含字母、数字、符号和空格等字符。

# 字符串的创建
# 你可以使用单引号 ' 或双引号 " 来定义一个字符串。如果字符串跨越多行，可以使用三重引号 ''' 或 """。

# 使用单引号定义字符串
single_quote_str = 'Hello, World!'

# 使用双引号定义字符串
double_quote_str = "Hello, World!"

# 使用三重引号定义多行字符串
multi_line_str = '''This is a multi-line string.
It spans multiple lines.'''

# 使用三重引号定义多行字符串
multi_line_str_2 = """This is another multi-line string.
It also spans multiple lines."""

#访问字符串中的字符
# 你可以使用索引来访问字符串中的字符。索引从0开始，负索引表示从字符串末尾开始计数。

# 定义一个字符串
greeting = "Hello, World!"

# 访问字符串中的第一个字符
first_char = greeting[0]  # 'H'

# 访问字符串中的最后一个字符
last_char = greeting[-1]  # '!'

# 访问字符串中的一个子串
substring = greeting[7:12]  # 'World'

# 字符串的不可变性
# 字符串是不可变的，这意味着你不能修改字符串中的字符。如果尝试修改字符串中的字符，会引发错误。

# 尝试修改字符串中的字符（会引发错误）
# greeting[0] = 'h'  # TypeError: 'str' object does not support item assignment

# 字符串的常用操作
# 尽管字符串是不可变的，但你仍然可以进行一些常见的操作，如连接、重复和切片。

# 定义两个字符串
str1 = "Hello"
str2 = "World"

# 连接字符串
concatenated_str = str1 + ", " + str2 + "!"
print(concatenated_str)  # 'Hello, World!'

# 重复字符串
repeated_str = str1 * 3
print(repeated_str)  # 'HelloHelloHello'

# 切片字符串
sliced_str = concatenated_str[7:12]
print(sliced_str)  # 'World'

# 字符串的常用方法
# Python的字符串提供了许多内置方法，以下是一些常用的方法：

# 定义一个字符串
text = "Hello, World!"

# 使用 len() 函数获取字符串的长度
length = len(text)  # 13

# 使用 lower() 方法将字符串转换为小写
lower_text = text.lower()
print(lower_text)  # 'hello, world!'

# 使用 upper() 方法将字符串转换为大写
upper_text = text.upper()
print(upper_text)  # 'HELLO, WORLD!'

# 使用 split() 方法将字符串拆分为列表
words = text.split(", ")
print(words)  # ['Hello', 'World!']

# 使用 join() 方法将列表连接为字符串
joined_text = " ".join(words)
print(joined_text)  # 'Hello World!'

# 使用 replace() 方法替换字符串中的子串
replaced_text = text.replace("World", "Python")
print(replaced_text)  # 'Hello, Python!'

# 使用 strip() 方法去除字符串两端的空白字符
whitespace_str = "   Hello, World!   "
stripped_str = whitespace_str.strip()
print(stripped_str)  # 'Hello, World!'

# 字符串的格式化
# Python提供了多种字符串格式化的方法，使得字符串的拼接和插值更加方便。

# 使用百分号 % 进行格式化
name = "Alice"
age = 30
formatted_str = "My name is %s and I am %d years old." % (name, age)
print(formatted_str)  # 'My name is Alice and I am 30 years old.'

# 使用 str.format() 方法
formatted_str = "My name is {} and I am {} years old.".format(name, age)
print(formatted_str)  # 'My name is Alice and I am 30 years old.'

# 使用 f-string（格式化字符串）
formatted_str = f"My name is {name} and I am {age} years old."
print(formatted_str)  # 'My name is Alice and I am 30 years old.'
```

- ### 字典

```python
# 字典的定义
# 在Python中，字典（dictionary）是一种用于存储键值对（key-value pairs）的数据结构。
# 字典是无序的、可变的，并且键必须是唯一的。
# 字典通常用于快速查找和存储数据。

# 字典的创建
# 你可以使用花括号 {} 来定义一个字典，并用冒号 : 分隔键和值，用逗号 , 分隔每个键值对。
# 以下是一些定义字典的示例：

# 定义一个空字典
empty_dict = {}

# 定义一个包含整数键和值的字典
numbers = {1: 'one', 2: 'two', 3: 'three'}

# 定义一个包含字符串键和值的字典
fruits = {'apple': 1, 'banana': 2, 'cherry': 3}

# 定义一个包含不同数据类型的字典
mixed_dict = {'name': 'Alice', 'age': 30, 'is_student': False}

# 定义一个嵌套字典（字典中的值也是字典）
nested_dict = {'person1': {'name': 'Alice', 'age': 30}, 'person2': {'name': 'Bob', 'age': 25}}

# 访问字典中的值
# 你可以使用键来访问字典中的值。如果键不存在，会引发 KeyError 异常。

# 定义一个字典
fruits = {'apple': 1, 'banana': 2, 'cherry': 3}

# 访问字典中的值
apple_count = fruits['apple']  # 1

# 使用 get() 方法访问字典中的值（如果键不存在，返回 None 或指定的默认值）
banana_count = fruits.get('banana')  # 2
orange_count = fruits.get('orange', 0)  # 0

# 修改字典中的值
# 你可以通过键来修改字典中的值。如果键不存在，会添加一个新的键值对。

# 修改字典中的值
fruits['banana'] = 5
print(fruits)  # {'apple': 1, 'banana': 5, 'cherry': 3}

# 添加新的键值对
fruits['orange'] = 4
print(fruits)  # {'apple': 1, 'banana': 5, 'cherry': 3, 'orange': 4}

# 删除字典中的键值对
# 你可以使用 del 关键字或 pop() 方法删除字典中的键值对。

# 使用 del 关键字删除键值对
del fruits['cherry']
print(fruits)  # {'apple': 1, 'banana': 5, 'orange': 4}

# 使用 pop() 方法删除键值对，并返回被删除的值
banana_count = fruits.pop('banana')
print(banana_count)  # 5
print(fruits)  # {'apple': 1, 'orange': 4}

# 使用 clear() 方法清空字典
fruits.clear()
print(fruits)  # {}

# 字典的常用方法
# Python的字典提供了许多内置方法，以下是一些常用的方法：

# 定义一个字典
person = {'name': 'Alice', 'age': 30, 'city': 'New York'}

# 使用 keys() 方法获取所有键
keys = person.keys()
print(keys)  # dict_keys(['name', 'age', 'city'])

# 使用 values() 方法获取所有值
values = person.values()
print(values)  # dict_values(['Alice', 30, 'New York'])

# 使用 items() 方法获取所有键值对
items = person.items()
print(items)  # dict_items([('name', 'Alice'), ('age', 30), ('city', 'New York')])

# 使用 update() 方法更新字典
person.update({'age': 31, 'city': 'San Francisco'})
print(person)  # {'name': 'Alice', 'age': 31, 'city': 'San Francisco'}

# 使用 copy() 方法复制字典
person_copy = person.copy()
print(person_copy)  # {'name': 'Alice', 'age': 31, 'city': 'San Francisco'}

# 字典的遍历
# 你可以使用 for 循环遍历字典的键、值或键值对。

# 定义一个字典
fruits = {'apple': 1, 'banana': 2, 'cherry': 3}

# 遍历字典的键
for key in fruits.keys():
    print(key)

# 遍历字典的值
for value in fruits.values():
    print(value)

# 遍历字典的键值对
for key, value in fruits.items():
    print(f"{key}: {value}")
```

- ### 集合

```python
# 集合的定义
# 在Python中，集合（set）是一种用于存储多个不重复元素的无序集合。
# 集合中的元素是唯一的，这意味着集合中不会出现重复的元素。
# 集合主要用于去重和集合运算（如并集、交集、差集等）。

# 集合的创建
# 你可以使用花括号 {} 或 set() 函数来定义一个集合。以下是一些定义集合的示例：

# 使用花括号定义一个空集合（注意：{} 定义的是空字典）
empty_set = set()

# 使用花括号定义一个包含整数的集合
numbers = {1, 2, 3, 4, 5}

# 使用花括号定义一个包含字符串的集合
fruits = {"apple", "banana", "cherry"}

# 使用 set() 函数定义一个集合
mixed_set = set([1, "hello", 3.14, True])

# 定义一个嵌套集合（集合中的元素不能是可变类型，如列表）
# nested_set = {{1, 2}, {3, 4}}  # 会引发错误

# 访问集合中的元素
# 由于集合是无序的，因此不能使用索引来访问集合中的元素。不过，你可以使用 for 循环遍历集合中的元素。

# 定义一个集合
fruits = {"apple", "banana", "cherry"}

# 遍历集合中的元素
for fruit in fruits:
    print(fruit)

# 修改集合中的元素
# 集合是可变的，这意味着你可以添加或删除集合中的元素。

# 定义一个集合
fruits = {"apple", "banana", "cherry"}

# 使用 add() 方法添加元素
fruits.add("orange")
print(fruits)  # {"apple", "banana", "cherry", "orange"}

# 使用 remove() 方法删除元素（如果元素不存在，会引发 KeyError 异常）
fruits.remove("banana")
print(fruits)  # {"apple", "cherry", "orange"}

# 使用 discard() 方法删除元素（如果元素不存在，不会引发异常）
fruits.discard("banana")
print(fruits)  # {"apple", "cherry", "orange"}

# 使用 pop() 方法删除并返回一个随机元素
random_fruit = fruits.pop()
print(random_fruit)
print(fruits)

# 使用 clear() 方法清空集合
fruits.clear()
print(fruits)  # set()

# 集合的常用操作
# 集合支持许多常见的集合运算，如并集、交集、差集和对称差集。

# 定义两个集合
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# 并集（union）
union_set = set1.union(set2)
print(union_set)  # {1, 2, 3, 4, 5, 6, 7, 8}

# 交集（intersection）
intersection_set = set1.intersection(set2)
print(intersection_set)  # {4, 5}

# 差集（difference）
difference_set = set1.difference(set2)
print(difference_set)  # {1, 2, 3}

# 对称差集（symmetric_difference）
symmetric_difference_set = set1.symmetric_difference(set2)
print(symmetric_difference_set)  # {1, 2, 3, 6, 7, 8}

# 集合的常用方法
# Python的集合提供了许多内置方法，以下是一些常用的方法：

# 定义一个集合
numbers = {1, 2, 3, 4, 5}

# 使用 len() 函数获取集合的长度
length = len(numbers)  # 5

# 使用 in 运算符检查元素是否在集合中
is_in_set = 3 in numbers  # True
is_not_in_set = 6 in numbers  # False

# 使用 copy() 方法复制集合
numbers_copy = numbers.copy()
print(numbers_copy)  # {1, 2, 3, 4, 5}

# 使用 update() 方法将另一个集合的元素添加到当前集合
numbers.update({6, 7})
print(numbers)  # {1, 2, 3, 4, 5, 6, 7}

# 使用 issubset() 方法检查一个集合是否是另一个集合的子集
is_subset = {1, 2}.issubset(numbers)  # True

# 使用 issuperset() 方法检查一个集合是否是另一个集合的超集
is_superset = numbers.issuperset({1, 2})  # True
```

- ### 实践指导

```python
# 练习数据结构的操作

# 1. 统计单词出现的频率
# 定义一个字符串
text = "This is a test string. This string is used for testing purposes."
# 将字符串按空格分割成单词列表
words = text.split()
# 创建一个空字典，用于存储单词和对应的频率
word_freq = {}
# 遍历单词列表
for word in words:
    # 如果单词已经在字典中，则将其频率加1
    if word in word_freq:
        word_freq[word] += 1
    # 如果单词不在字典中，则将其添加到字典，并将其频率设置为1
    else:
        word_freq[word] = 1
# 打印每个单词及其频率
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

```python
# 2. 学生成绩管理系统
# 创建学生列表
students = []

# 添加学生信息
students.append({"姓名": "张三", "成绩": 90})
students.append({"姓名": "李四", "成绩": 85})
students.append({"姓名": "王五", "成绩": 95})

# 显示学生信息
for student in students:
    print("姓名:", student["姓名"])
    print("成绩:", student["成绩"])
    print()

# 计算平均成绩
total_score = 0
for student in students:
    total_score += student["成绩"]
average_score = total_score / len(students)
print("平均成绩:", average_score)
```
